```python
import pygame, sys, time, random

#difficulty
difficulty = 10

level = 1  #starting level
score_to_next_level = 4  #points needed to level up

#window size
frame_size_x = 720
frame_size_y = 480

#initialize pygame
pygame.init()
pygame.display.set_caption('snake')
game_window = pygame.display.set_mode((frame_size_x, frame_size_y))

#define colors
black = pygame.Color(0, 0, 0)
white = pygame.Color(255, 255, 255)
red = pygame.Color(255, 0, 0)
green = pygame.Color(0, 255, 0)

#fps controller
fps_controller = pygame.time.Clock()

#snake initial position and body
snake_pos = [100, 50]
snake_body = [[100, 50], [90, 50], [80, 50]]

#function to generate food position(not on the snake)
def spawn_food():
    while True:
        food_x = random.randint(0, (frame_size_x // 10) - 1) * 10
        food_y = random.randint(0, (frame_size_y // 10) - 1) * 10
        if [food_x, food_y] not in snake_body:
            return [food_x, food_y]

food_pos = spawn_food()

direction = 'RIGHT'
change_to = direction
score = 0

#function to handle game over
def game_over():
    my_font = pygame.font.SysFont('times new roman', 90)
    game_over_surface = my_font.render('game over', True, red)
    game_over_rect = game_over_surface.get_rect(midtop=(frame_size_x/2, frame_size_y/4))
    game_window.fill(black)
    game_window.blit(game_over_surface, game_over_rect)
    show_score_level()
    pygame.display.flip()
    time.sleep(3)
    pygame.quit()
    sys.exit()

#display score and level
def show_score_level():
    score_font = pygame.font.SysFont('consolas', 20)
    score_surface = score_font.render(f'score: {score}  level: {level}', True, white)
    score_rect = score_surface.get_rect(midtop=(frame_size_x/2, 15))
    game_window.blit(score_surface, score_rect)

#main game loop
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP and direction != 'DOWN':
                change_to = 'UP'
            if event.key == pygame.K_DOWN and direction != 'UP':
                change_to = 'DOWN'
            if event.key == pygame.K_LEFT and direction != 'RIGHT':
                change_to = 'LEFT'
            if event.key == pygame.K_RIGHT and direction != 'LEFT':
                change_to = 'RIGHT'
            if event.key == pygame.K_ESCAPE:
                pygame.event.post(pygame.event.Event(pygame.QUIT))

    #update direction
    direction = change_to

    #move the snake
    if direction == 'UP':
        snake_pos[1] -= 10
    elif direction == 'DOWN':
        snake_pos[1] += 10
    elif direction == 'LEFT':
        snake_pos[0] -= 10
    elif direction == 'RIGHT':
        snake_pos[0] += 10

    #snake growth logic
    snake_body.insert(0, list(snake_pos))
    if snake_pos == food_pos:
        score += 1
        food_pos = spawn_food()
        if score % score_to_next_level == 0:
            level += 1
            difficulty += 5
    else:
        snake_body.pop()

    #check for collisions with walls
    if snake_pos[0] < 0 or snake_pos[0] >= frame_size_x or snake_pos[1] < 0 or snake_pos[1] >= frame_size_y:
        game_over()
    
    #check for collisions with itself
    if snake_pos in snake_body[1:]:
        game_over()

    #draw game elements
    game_window.fill(black)
    for pos in snake_body:
        pygame.draw.rect(game_window, green, pygame.Rect(pos[0], pos[1], 10, 10))
    pygame.draw.rect(game_window, white, pygame.Rect(food_pos[0], food_pos[1], 10, 10))

    #display score and level
    show_score_level()

    #update display
    pygame.display.update()

    #set game speed
    fps_controller.tick(difficulty)
```
